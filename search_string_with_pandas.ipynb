# pandasの文字列検索

```python
import numpy as np
import pandas as pd
```

## データ生成

```python
sample_data = np.array([
    ["aaa", "bbb", "ccc0",    "ddd"],
    ["aaa", "bbb", "ccc1",    "ddd"],
    ["aaa", "bbb", "cccxxx2", "ddd"],
    ["aaa", "bbb", "ccc3",    "ddd"],
    ["aaa", "bbb", "cccxxx4", "ddd"],
])
```

```python
col = ["AAA", "BBB", "CCC", "DDD"]
df = pd.DataFrame(data=sample_data,
                  columns=col)
df
```

## "xxx"の検索

```python
xxxIdx = df[df["CCC"].str.contains("xxx")].index
print(xxxIdx)
print(type(xxxIdx))
```

## "xxx"の一行前の列の抽出

```python
xxxIdx - 1
```

```python
df.loc[xxxIdx-1, :]
```

```python
df.loc[xxxIdx-1, ["BBB", "CCC"]]
```
